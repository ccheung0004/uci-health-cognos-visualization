```python
import pandas as pd

```

```python

#The first file has an error.
food_summary_files = [
    'Andriola daily servings summary 190701 thru 191231.csv',
    'Andriola daily servings summary 200101 thru 201231.csv',
    'Andriola daily servings summary 210101 thru 211231.csv',
    'Andriola daily servings summary 220101 thru 220615.csv'
]
```

```python
#load data
```

```python
cols = pd.read_csv(food_summary_files[0]).columns;
dfs = []
for summary_file in food_summary_files:
    df = pd.read_csv(summary_file)
    dfs.append(df)
food_df = pd.concat(dfs, ignore_index=True).dropna()
```

```python
#process data
```

```python
pattern = r'(\d{4})-(\d{2})-(\d{2})'
replacement = r'\2/\3/\1'
food_df['Day'] = food_df['Day'].replace(pattern, replacement, regex=True)
food_df['ts'] = food_df['Day'] + ' ' + food_df['Time']
food_df['ts'] = pd.to_datetime(food_df['ts'])
```

```python
#analysis of food groups
food_df['Group'].value_counts()
```

```python
#Analysis of food items
food_df['Food Name'].value_counts()
```

```python
arboleaf_data_file = 'arboleaf-Tom Andriola 2020-01 to 2022-06.csv'
arboleaf_df=  pd.read_csv(arboleaf_data_file)
```

```python
#Parse the data
arboleaf_df['ts'] = pd.to_datetime(arboleaf_df['Time of Measurement'].str.replace("，",","), format='%B %d, %Y at %I:%M:%S %p').dt.date
```

```python
sleep_data_file = "oura_2020-04-14_2022-06-11_trends.csv"
sleep_df = pd.read_csv(sleep_data_file)
```

```python
sleep_df.columns
```

```python
#parse the data
sleep_df['date'] = pd.to_datetime(sleep_df['date'])
```

```python
sleep_df
```

```python
# First join sleep and arboleaf together by the day.
arboleaf_cols = []
arboleaf_by_day = arboleaf_df[['ts', 'Weight(lb)', 'BMI', 'Body Fat(%)', 'Fat-free Body Weight(lb)', 'Subcutaneous Fat(%)', 'Visceral Fat',
       'Body Water(%)', 'Skeletal Muscle(%)', 'Muscle Mass(lb)',
       'Bone Mass(lb)', 'Protein(%)',
       'BMR(kcal)', 'Metabolic Age']].groupby('ts').mean()  # Turn into averages for that day
arboleaf_by_day.index = pd.to_datetime(arboleaf_by_day.index)

# Use outer join to merge, including all records from both dataframes
merged_sleep_and_bio = pd.merge(arboleaf_by_day, sleep_df, left_index=True, right_on="date", how='outer')

# Setting the index
merged_sleep_and_bio.set_index('date', inplace=True)
```

```python
merged_sleep_and_bio
```

```python
food_df['day_ts'] = food_df['ts'].dt.date
ignore = set(['Food Name','Day','Time','Group','Category','ts','Amount'])
food_df_by_day = food_df[list(set(food_df.columns) - ignore)]
food_df_by_day.set_index('day_ts', inplace=True)
food_df_by_day = food_df_by_day.groupby('day_ts').sum()
```

```python
merged_bio_food = pd.merge(food_df_by_day, merged_sleep_and_bio, left_index=True, right_index=True, how = 'outer')
```
